```python
# All filenames are in the following format:
# SQUARE_WAVE_nxf_0.csv
# For n = 1, 3, 5, 7, 9

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read all of the files
data = []
for i in range(1, 10, 2):
    data.append(pd.read_csv(f"SQUARE_WAVE_{i}xf_0.csv", header=None))

# Change the first row of each file
for i in range(5):
    data[i].iloc[0, 0] = "Time (ms)"
    data[i].iloc[0, 1] = "Channel A (V)"
    data[i].iloc[0, 2] = "Channel B (V)"
    # Get rid of the 2nd row
    data[i] = data[i].drop(1)

# Convert entries to float
for i in range(5):
    data[i].iloc[1:, 0] = data[i].iloc[1:, 0].astype(float)
    data[i].iloc[1:, 1] = data[i].iloc[1:, 1].astype(float)
    data[i].iloc[1:, 2] = data[i].iloc[1:, 2].astype(float)

# Get head of all the files
for i in range(5):
    print(f"File {i+1}")
    print(data[i].head())
    print()

```

```python
# Plot the first 1000 rows of data
plt.figure()
for i in range(5):
    plt.plot(data[i].iloc[1:, 0], data[i].iloc[1:, 2], label=f"Channel B n = {2*i + 1}")
plt.ylim(-10, 10)
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (V)")
plt.title("Square Waveforms")
plt.legend()
plt.show()
```
